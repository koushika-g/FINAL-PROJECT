```python
#scrapping with selenium 

import selenium
from selenium import webdriver 
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
import os
from selenium.webdriver.common.by import By
```

```python
driverpath=r"C:\Users\Merit\Downloads\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverpath)
driver.maximize_window()
driver.get("https://www.cliffordchance.com/people_and_places.html?_charset_=UTF-8&fname=&lname=&tags=all&tags=all&office=%2Fcontent%2Fcliffordchance%2Fpeople_and_places%2Fpeople%2Foffices%2Fluxembourg&x=64&y=19&pageitems=50&counselsview=true#counsels")

driver.implicitly_wait(10)
time.sleep(5)
driver.find_element(By.XPATH, '//a[text()="OK"]').click()

name=[]
des=[]
loc=[]
phn=[]
p=[]

#dropdown

dropd=driver.find_element_by_xpath("//select[@id='lawyerssearchnum']")
driver.execute_script("arguments[0].click();",dropd)
downd=driver.find_element_by_xpath("//header/select[@id='lawyerssearchnum']/option[3]")
driver.execute_script("arguments[0].click();",downd)

category=[]
cat=driver.find_elements_by_xpath("//a[@class='icon_profile']")
for m in cat:
    links=m.get_attribute("href")
    category.append(links)
#     print(category)

#fetching all details
for s in category:
    driver.get(s)
    
    #name
    a=driver.find_elements_by_xpath("//h2[contains(@class,'utils_black')]")
    for i in a:
        name.append(i.text)
        print(name)
        
    #role
    b=driver.find_elements_by_xpath("//h3[contains(@class,'utils_black')]")
    for j in b:
        des.append(j.text)
        print(des)
        
    #location
    c=driver.find_elements_by_xpath("//h2[contains(text(),'Contact details')]//following::a[1]")
    for k in c:
        loc.append(k.text)
        print(loc)
        
    #phone number
    d=driver.find_elements_by_xpath("//h2[contains(text(),'Contact details')]//following::li[2]")
    for l in d:
        if "+" in l.text:
            phn.append(l.text)
            print(phn)
        else:
            phn.append("null")
        
    #url link
    p.append(s)
    
    driver.back()
    time.sleep(2)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#dataframe
df = pd.DataFrame()
df['SNO'] = df.index
df.set_index('SNO',inplace=True)
df['NAME'] = name
df['ROLE'] = des
df['LOCATION'] = loc
df['PHONENUMBER'] = phn
df['URL'] = p
df['DATE TIME']= pd.Timestamp(datetime.now())
df
```

```python
#flaskAPI with SQLlite

import os
import sqlite3
from flask import jsonify
from flask_api import FlaskAPI
from flask_restful import Api
from sqlalchemy import create_engine
from flask import request
import json

#default route 

app=FlaskAPI(__name__)
api=Api(app)
eng=create_engine('sqlite:///clifford.db') 

#create

@app.route('/cliffordchance_create', methods=['GET'])
def create():
    try:
        con = eng.connect()
        trans = con.begin()
        con.execute('CREATE TABLE IF NOT EXISTS datatbl(SNO INT PRIMARY KEY,NAME TEXT,ROLE TEXT,LOCATION TEXT,PHONENUMBER INT,URLS TEXT,Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP)')
        trans.commit()
        text = {"Message " : "success"}
        return text
    except Exception as error:
        text = {"Message " : "Failed"}
        return text
    finally:
        con.close()

#insert

@app.route('/cliffordchance_insert', methods=['GET'])
def insert_tbl():
    con = eng.connect()
    df.to_sql('datatbl', con,if_exists='replace')
    text = {"Message " : "Inserted Successfully"}
    return text

#select

@app.route('/cliffordchance_select', methods=['GET'])
def select():
    con = eng.connect()
    data = con.execute("select * from datatbl")  
    return json.dumps([dict(s) for s in data])
    
app.run()
```
